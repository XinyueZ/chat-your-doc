```python
from typing import Any
from rich.pretty import pprint


def pretty_print(title: str = None, content: Any = None):
    if title is None:
        print(content)
        return
    print(title)
    pprint(content)
```

```python
from llama_index.llms.huggingface import HuggingFaceLLM
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch


# https://github.com/chentong0/factoid-wiki
"""
def test_model(model, tokenizer, device):
    import json

    title = "Leaning Tower of Pisa"
    section = ""
    content = "Prior to restoration work performed between 1990 and 2001, Leaning Tower of Pisa leaned at an angle of 5.5 degrees, but the tower now leans at about 3.99 degrees. This means the top of the tower is displaced horizontally 3.9 meters (12 ft 10 in) from the center."

    input_text = f"Title: {title}. Section: {section}. Content: {content}"

    input_ids = tokenizer(input_text, return_tensors="pt").input_ids
    outputs = model.generate(input_ids.to(device), max_new_tokens=512).cpu()

    output_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    try:
        prop_list = json.loads(output_text)
    except:
        prop_list = []
        print("[ERROR] Failed to parse output text as JSON.")
    print(json.dumps(prop_list, indent=2))
"""

hf_model_name = "chentong00/propositionizer-wiki-flan-t5-large"
device = "cuda" if torch.cuda.is_available() else "cpu"
hf_tokenizer = AutoTokenizer.from_pretrained(hf_model_name, use_fast=False)
hf_tokenizer.model_max_length = 2048
hf_model = AutoModelForSeq2SeqLM.from_pretrained(hf_model_name).to(device)

# test_model(hf_model, hf_tokenizer, device)


proposition_llm = HuggingFaceLLM(
    tokenizer=hf_tokenizer,
    model=hf_model,
    device_map=device,
)
```

#### Any model is possible

Here uses free ones.

```python
from llama_index.llms.groq import Groq
query_llm = Groq(model="mixtral-8x7b-32768", api_key=groq_api_key)

from llama_index.legacy.embeddings.langchain import LangchainEmbedding
from langchain_nvidia_ai_endpoints import NVIDIAEmbeddings
nv_embedding = NVIDIAEmbeddings(model="nvolveqa_40k", nvidia_api_key=nvidia_api_key)
embed_model=LangchainEmbedding(nv_embedding)
```

```python
!llamaindex-cli download-llamapack DenseXRetrievalPack --download-dir ./dense_pack
```

```python
from llama_index.core import SimpleDirectoryReader
documents = SimpleDirectoryReader(input_files=[FILE_NAME]).load_data()
```

#### Consturct the DenseXRetrieval

It will take a few minutes as the proposition model runs locally.

```python
from dense_pack.llama_index.packs.dense_x_retrieval.base import DenseXRetrievalPack
dense_pack = DenseXRetrievalPack(
    documents,
    proposition_llm=proposition_llm,
    query_llm=query_llm,
    embed_model=embed_model,
)
```

```python
response = dense_pack.run("What is the content about?")

pretty_print("response.response", response.response )
```
